```python
import pandas as pd
from datetime import datetime


```

```python
# Carregamento do arquivo do arquivo
df = pd.read_excel('/content/Teste.xlsx')

```

```python
display(df)
```

```python
# Conversão da coluna 'data_compra' para o tipo de dados datetime
df['data_compra'] = pd.to_datetime(df['data_compra'])

```

```python
# Classificação do DataFrame pelo nome do comprador e pela data da compra
df.sort_values(by=['nome_comprador', 'data_compra'], inplace=True)

```

```python
# Cálculo da diferença de tempo entre compras consecutivas para cada cliente
df['tempo_entre_compras'] = df.groupby('nome_comprador')['data_compra'].diff()

```

```python
# Impressão
display(df)
```

```python
# Cálculo da média do tempo de retorno para todos os clientes
tempo_medio_retorno = df['tempo_entre_compras'].mean()

```

```python
display(f'O tempo médio de retorno dos clientes é: {tempo_medio_retorno}')
```

```python
# Criar uma tabela de frequência para o número de retornos
retorno_frequencia = df.groupby('nome_comprador').size().reset_index(name='Numero_de_Retornos')

```

```python
# Adicionar uma coluna para a média do tempo de retorno em cada linha
retorno_frequencia['Media_Tempo_Retorno'] = df.groupby('nome_comprador')['tempo_entre_compras'].mean().values
```

```python
# Crosstab para criar uma matriz cruzada
matriz_cruzada = pd.crosstab(retorno_frequencia['Numero_de_Retornos'], columns='count', values=retorno_frequencia['Media_Tempo_Retorno'], aggfunc='mean')
```

```python
print(f'O tempo médio de retorno dos clientes é: {tempo_medio_retorno}')
display(matriz_cruzada)
```
